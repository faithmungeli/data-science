# DATA CLEANING

# get mean of age

```python
mean_Age=train['Age'].mean()
```

```python
mean_Age
```

```python
train['Age']= train['Age'].fillna((train['Age'].mean()))
```

```python
train['Age'].isnull().sum()
```

```python
def fill_age(col):
    Age=col[0]
    if pd.isnull(Age):
        return mean_age
    else:
        return Age
```

# call age column and apply the mean age

```python
train['Age']=train[['Age']].apply(fill_age,axis=1)
```

# visualize missing values

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

# drop cabin data

```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

# Machine learning

# converting catgorical features 

```python
sex=pd.get_dummies(train['Sex'])
```

```python
sex
```

```python
sex=pd.get_dummies(train['Sex'],drop_first=True)
```

```python
sex
```

# convert the embark into categorical or binary feature

# c=cherbourg, Q=queenstown S=southampton

```python
embark=pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
embark
```

# drop the columns not needed.(Sex,name,embarked,ticket)

```python
train.head(5)
```

```python
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
train.head()
```

# Add the newly created columns after categorical features are added

```python
train=pd.concat([train,sex,embark],axis=1)
```

```python
train.head()
```

# split dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(train.drop('Survived',axis=1),train['Survived'],test_size=0.30,random_state=101)
```

# creating our model

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

# prediction

```python
prediction=logmodel.predict(x_test)
```

# metrics which are recall,precision,f-measure

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test,prediction))
```

```python
cm=(confusion_matrix(y_test,prediction))
```

```python
cm
```

```python
tn,fp,fn,tp=confusion_matrix(y_test,prediction).ravel()
```

```python
print((tp+tn)/(tn+fp+fn+tp))
```

```python
accuracy_score(y_test,prediction)
```

# SVM-(support vector machine)
#it works well in textual data and maximizes the actual hyperplane
#it creates a divider as opposed to linear regression and the parameters will determine the points to draw the support vectors.
#svm does maximum margin divide by 2

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
prediction2=model.predict(x_test)
```

```python
print(confusion_matrix(y_test,prediction2))
```

```python
accuracy_score(y_test,prediction2)
```

# GRID SEARCH -finding the right paramaters
it describes th e paramaters that should be tried and a model train.
it is the idea of creating grid paramaters and trying out  all possible combinations.

```python
param_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.0001,],'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=7)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions=grid.predict(x_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
accuracy_score(y_test,grid_predictions)
```
